```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
chi2 = stats.chi2(df = 2)
x = np.linspace(0, 12, 1_000)


fig, ax = plt.subplots()
for spine in 'left', 'right', 'top':
    ax.spines[spine].set_visible(False)

ax.plot(x, chi2.pdf(x))

# Lables
ax.set_title(r"$\chi^2$ curve, 2 degrees of freedom")

ax.axvline(5, ls = 'dashed', color = 'gray')

right = x[x>=5]
ax.fill_between(right, chi2.pdf(right), color = 'C1', alpha = 0.8)
ax.text(6, chi2.pdf(6) + .01, 
       s = 'P = 0.082',
       ha = 'left',
       va = 'bottom')

ax.text(5, chi2.pdf(5)*5, 
       s = r' $\chi^2$-statistic',
       ha = 'left',
       va = 'center')

ylims = ax.get_ylim()
ax.set_ylim(0, ylims[1])
ax.set_yticks([])

plt.tight_layout()
plt.savefig("../book/images/chi2Pval.svg", transparent = True)
```
