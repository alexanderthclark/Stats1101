# Histograms

```python
import matplotlib.pyplot as plt
```

```python
data = [1,1,1,1,2,3,3,3,4,4,100]
plt.hist(data)
plt.show()
```

```python
data = [1,1,1,1,2,3,3,3,4,4,100]
plt.hist(data, bins = 33)
plt.show()
```

```python
data = [1,1,1,1,2,3,3,3,4,4,100]
plt.hist(data, bins = [0,50,99,102])
plt.show()
```
