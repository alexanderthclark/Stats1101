```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
np.random.seed(1)
```

```python
data = stats.norm.rvs(size = 10_000)

avg = np.mean(data)
med = np.median(data)

fig, ax = plt.subplots()
ax.axis('off')
ax.hist(data, ec = 'black')

ax.axvline(avg, ls = 'dashed', color = 'C1', lw = 4, label = 'average')
ax.axvline(med, ls = 'dashed', color = 'white', lw = 2, label = 'median')

l = ax.legend()
l.get_frame().set_facecolor('lightgray')
l.get_frame().set_edgecolor('black')

plt.savefig("hist_with_mean_median.svg", transparent = True, format = 'svg')
plt.show()
```
